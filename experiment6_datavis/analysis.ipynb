```python
import json
import matplotlib.pyplot as plt

# Load the data
with open('linguistic.json', 'r') as f:
    linguistic_data = json.load(f)
with open('reported.json', 'r') as f:
    reported_data = json.load(f)

# Calculate the match percentages
linguistic_match_percentage = sum(1 for item in linguistic_data if item['prediction'] == item['truth']) / len(linguistic_data) * 100
reported_match_percentage = sum(1 for item in reported_data if item['prediction'] == item['truth']) / len(reported_data) * 100

# Create the bar graph
plt.bar(['Linguistic', 'Reported'], [linguistic_match_percentage, reported_match_percentage])
plt.ylabel('Match Percentage')
plt.title('Match Percentage of Predictions and Truth Labels')
plt.show()
```
